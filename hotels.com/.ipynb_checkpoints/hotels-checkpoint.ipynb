```python
file1 = open('mountains.txt','r')
mountain_names = set([x.replace('\n','') for x in file1.readlines()])
chains_names = set(['A&O',"AC Hotels","Adagio","Apollo","Arcadia","Ascot","Atahotels","Autograph","Baglioni","Belmond","Best Western","Blue Hotels","BlueBay","Boscolo","BQ","Britannia","Bulgari","Campanile","Charme","Chateaux","Clarion","Comfort","Concorde","Country Inn","Courtyard","Crowne","Days Inn","Desing Hotels","Diamond","Domina","Dorchester","Doubletree","Europa","Eurostars","Falkensteiner","Floris","Four Points","Four Seasons","Golden Tulip","Gran Meli","Grange","H10","Hampton","Hilton","Holiday Inn","Elba","Hotusa","Hyatt","ibis","Idea","Indigo","InterContinental","Marriot","Kyriad","Land-gut","Landidyll","Le Meridien","Leonardo","Logis","Mandarin","Meininger","Meliá","Mercure","MGallery","Milennium","Minotel","Montana","NH Hotel","Nhow","Novotel","Oceania","Park Inn","Pierre & Vacances","Preffered","Pullman","Quality","Qualys","Radisson","Ramada","Regent","Relais","Rinaldi","Robinson","Rocco","Romantik","Sea Hotels","Seaside","Sheraton","Sina","Sleep Inn","Sofitel","St. Regis","Star Inn","Starhotels","Starwood","Swiss Quality","The Luxury","TOP Hotel","Tulip Inn","Una Hotels","UNAWAY","Warwick","Westin","Worldhotels","Wyndham"])
```

```python
all_df = pd.DataFrame(columns=["Name of Hotel","Adress","City","Number of Stars","Things to do","Working Away","Mountains","Beaches","Chain Y/N","Number of rooms","Price for night","Link"])
```

```python
loc = (r"C:\Users\saxou\Downloads\Scraping Hotel.xlsx")
wb = xlrd.open_workbook(loc) 
sheet = wb.sheet_by_index(0)
sheet.cell_value(0,0)
names = []
for i in range(sheet.nrows): 
    names.append(sheet.cell_value(i, 0))

names = list(dict.fromkeys(names))
#print(len(names))
    
names.remove(names[0])

names_1=names[0:400]
```

```python
links=[]
```

```python
len(links)
```

```python
names_1=names[225:400]
i=0
for i,name in enumerate(names_1):   
    try:
        search_area=driver.find_element_by_class_name("widget-query-autosuggest")
        search_area.send_keys(Keys.CONTROL, "a")
        search_area.send_keys(names_1[i]+" Italy")
    except:
        driver.find_element_by_class_name("widget-toggle-btn").click()
        time.sleep(1)
        search_area=driver.find_element_by_class_name("widget-query-autosuggest")
        search_area.send_keys(Keys.CONTROL, "a")
        search_area.send_keys(names_1[i]+" Italy")
        
    print(i)
    print(names_1[i])        
    time.sleep(2)
    
    try:
        a=driver.find_element_by_class_name("autosuggest-hotel")
        a.find_element_by_css_selector("td").click()
    except:
        pass
    
    search_button = driver.find_element_by_class_name("cta-strong").click()
    
    time.sleep(2)
    
    try:
        n = driver.find_element_by_class_name("p-name")
        if names_1[i].lower() in n.text.lower():
            print("yes")
            link = n.find_element_by_css_selector("a").get_attribute("href")
            links.append(link)
    except:
        title = driver.find_element_by_class_name("property-description")
        n = title.find_element_by_css_selector("h1")
        if names_1[i].lower() in n.text.lower():
            print("yes")
            link = driver.current_url
            links.append(link)     
    print(n.text)
    i+=1
```

```python
i=0
for i,link in enumerate(links):
    print(i)
    i+=1
    tmp={}
    driver.get(links[i])
    time.sleep(2)
    tmp['Link']=links[i]
    try:
        title = driver.find_element_by_class_name("property-description")
        name = title.find_element_by_css_selector("h1").text
    except:
        tmp["Name of Hotel"]=driver.find_element_by_class_name("p-name").text
        all_df = all_df.append(tmp,ignore_index=True)      
        i+=1
        print(i)
        driver.get(links[i])
        
    title = driver.find_element_by_class_name("property-description")
    name = title.find_element_by_css_selector("h1").text    
    #print(name)
    tmp['Name of Hotel']=name
    try:
        star_num = title.find_element_by_class_name("star-rating-text").text
        #print(star_num)
        tmp['Number of Stars']=star_num
    except:
        pass

    adress = title.find_element_by_class_name("postal-addr").text
    #print(adress)
    tmp['Adress']=adress
    
    #print(adress.split(",")[1].replace(",",""))
    if len(adress.split(",")[1].replace(",","")) >=5:
        tmp['City']=adress.split(",")[1].replace(",","")
    elif len(adress.split(",")[1].replace(",","")) <5:
        tmp['City']=adress.split(",")[2].replace(",","")
    
    info_box = driver.find_element_by_class_name("info-box")
    size = info_box.find_element_by_css_selector("li").text
    #print(size.split()[len(size.split())-2]+" "+size.split()[len(size.split())-1])
    tmp['Number of rooms']=size.split()[len(size.split())-2]+" "+size.split()[len(size.split())-1]
    
    
    rows = driver.find_elements_by_class_name("fact-sheet-table-row")
    for row in rows:
        header = row.find_element_by_class_name("fact-sheet-table-header").text
        if header == "Things to do":
            li = row.find_elements_by_css_selector("li")
            #print(len(li))
            tmp['Things to do']=len(li)
                              
    headers=[]
    rows = driver.find_elements_by_class_name("fact-sheet-table-row")
    for row in rows:
        header = row.find_element_by_class_name("fact-sheet-table-header").text       
        headers.append(header.lower())
    if any("working away" in s for s in headers):
        tmp["Working Away"]="1"
    else:
        tmp["Working Away"]="0"

        
    try:
        price = driver.find_element_by_class_name("current-price").text
        tmp['Price for night']=price
    except:
        pass
    
    try:
        lml = driver.find_element_by_class_name("landmark-list")
        lms = lml.find_elements_by_css_selector("li")
        for lm in lms:
            if "beach" in lm.text.lower():
                #print(lm.text)
                tmp['Beaches']=lm.text
            if "ski" in lm.text.lower():
                #print(lm.text)
                tmp['Mountains']=lm.text
    except:
        pass
            
    name=name.lower()
    tmp1={}
    for chn in chains_names:
        if chn.lower() in name:
            tmp1[chn] = len(chn)
    try:
        print(max(tmp1.items(), key=operator.itemgetter(1))[0])
        tmp['Chain Y/N']="Yes"
    except:
        tmp['Chain Y/N']="No"
    

    all_df = all_df.append(tmp,ignore_index=True)
```
