```python
all_df = pd.DataFrame(columns=["Title","Type","Link","City","Location","Number of Stars","Reviews","Food & Drink","Business Facilities","Entertainment & Family Services","Pool & Spa","Health & Wellness Facilities","Ski Activities","Nearby Beaches","Nearby Skilifts","Price"])
```

```python
hrefs = []
i=0
while i<16:
    hrefs_PE = driver.find_elements_by_class_name("hotel_name_link")
    for el in hrefs_PE:
        href = el.get_attribute("href")
        print(href)
        hrefs.append(href)
        nxt_PE=driver.find_element_by_class_name("bui-pagination__next-arrow")
        nxt=nxt_PE.find_element_by_css_selector("a").click()
        time.sleep(1)
        i+=1
```

```python
for href in hrefs:
    tmp = {}
    driver.get(href)  
    
    tmp['Link']=href
    
    SCROLL_PAUSE_TIME = 0.1 
    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    #Price
    try:
        price = driver.find_element_by_class_name("bui-price-display__value").text
        print(price)
        tmp['Price']=price
    except:
        try:
            time.sleep(2)
            price_PE = driver.find_element_by_class_name("availability_price")
            price = price_PE.find_element_by_css_selector("span").text
            tmp['Price']=price
        except:
            tmp["Price"]="-"

            
    #Title and Type
    title = driver.find_element_by_class_name("hp__hotel-name")
    #print(title.text)
    Type = title.find_element_by_css_selector("span").text
    tmp['Type']=Type
    #print(Type)
    if Type in title.text:
        tmp['Title']=title.text.replace(Type,"",1)

    
    
    #Location and City
    location = driver.find_element_by_class_name("hp_address_subtitle").text
    #print(location)
    tmp['Location']=location
    city = location.split()[len(location.split())-2].replace(",","")
    #print(city)
    tmp['City']=city
    
    #Reviews
    review_arr=[]
    try:
        reviews = driver.find_elements_by_class_name("c-review-snippet__review")
        for review in reviews:
            review = review.text
            if "Read more" in review or "..." in review or len(review)==0:
                pass
                #print("-")
            else:
                if len(review_arr)<4:
                    #print(review)
                    review_arr.append(review)
    except:
        driver.find_element_by_xpath('//*[@id="show_reviews_tab"]').click()
        time.sleep(1)
        reviews = driver.find_elements_by_class_name("c-review__body")    
        for review in reviews:
            review = review.text
            if "Read more" in review or "..." in review or len(review)==0:
                pass
                #print("-")
            else:
                if len(review_arr)<4:
                    #print(review)
                    review_arr.append(review)
    tmp['Reviews']="/".join(review_arr)
    
    #Star Track
    try:
        star_track = driver.find_element_by_class_name("star_track").get_attribute("title")        
        if "1-star" in star_track:
            tmp['Number of Stars']="1 star"
        elif "2-star" in star_track:
            tmp['Number of Stars']="2 star"
        elif "3-star" in star_track:
            tmp['Number of Stars']="3 star"
        elif "4-star" in star_track:
            tmp['Number of Stars']="4 star"
        elif "5-star" in star_track:
            tmp['Number of Stars']="5 star"
        else:
            tmp['Number of Stars']="-"
    except:
        tmp['Number of Stars']="-"

    #Facilities
    factis = driver.find_elements_by_class_name("facilitiesChecklistSection")
    for fact in factis:
        fact_title = fact.find_element_by_css_selector("h5").text
        #print(fact_title)
        if "Ski" in fact_title:
            tmp['Ski Activities']=fact.find_element_by_css_selector("ul").text.replace("Additional charge","",20)      
        if "Food & Drink" in fact_title:
            tmp['Food & Drink']=fact.find_element_by_css_selector("ul").text.replace("Additional charge","",20)
        if "Business Facilities" in fact_title:
            tmp['Business Facilities']=fact.find_element_by_css_selector("ul").text.replace("Additional charge","",20)
        if "Entertainment & Family Services" in fact_title:
            tmp['Entertainment & Family Services']=fact.find_element_by_css_selector("ul").text.replace("Additional charge","",20)
        if "Pool and Spa" in fact_title:
            tmp['Pool & Spa']=fact.find_element_by_css_selector("ul").text.replace("Additional charge","",20)
        if "Health & Wellness Facilities" in fact_title:
            tmp['Health & Wellness Facilities']=fact.find_element_by_css_selector("ul").text.replace("Additional charge","",20)
            
    #nearby skies and beaches
    try:
        nearby_beaches = driver.find_element_by_class_name("hp-closest-beaches").text
        tmp['Nearby Beaches']=nearby_beaches.lower().replace("popular beaches nearby","").replace("\n","",1)
    except:
        pass
    try:
        nearby_beaches = driver.find_element_by_class_name("hp-closest-skilifts").text
        tmp['Nearby Skilifts']=nearby_beaches.lower().replace("nearby ski lifts:","").replace("\n","",1)
    except:
        pass
    
    
    all_df = all_df.append(tmp,ignore_index=True)
    print(tmp)
```
