```python
hrefs = []
hrefs_PE = driver.find_elements_by_class_name("url")
for el in hrefs_PE:
    href = el.get_attribute("href")
    print(href)
    hrefs.append(href)
```

```python
prices = []
titles = []
locations = []
cities = []
reviewss = []
facilitiess = []


for href in hrefs:
    driver.get(href)
    SCROLL_PAUSE_TIME = 0.1
    try:
        driver.find_element_by_xpath('//*[@id="offcanvas-container"]/div[4]/div/div/a').click()
    except:
        pass    

    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
        
    try:
        price = driver.find_element_by_class_name("bui-price-display__value").text
        print(price)
        prices.append(price)
    except:
        try:
            time.sleep(1)
            price_PE = driver.find_element_by_class_name("availability_price")
            price = price_PE.find_element_by_css_selector("span").text
            prices.append(price)
        except:
            prices.append("-")
    
    title = driver.find_element_by_class_name("hp__hotel-name").text
    print(title)
    titles.append(title)
    
    location = driver.find_element_by_class_name("hp_address_subtitle").text
    print(location)
    locations.append(location)
    
    city = location.split()[len(location.split())-2].replace(",","")
    print(city)
    cities.append(city)
    
    driver.find_element_by_xpath('//*[@id="show_reviews_tab"]').click()
    time.sleep(1)
    reviews = driver.find_elements_by_class_name("c-review-block__title")
    #print(reviews)
    reviews_arr=[]
    for review in reviews:
        review = review.text
        if "Read more" in review or "..." in review:
            pass
            #print("-")
        else:
            if len(reviews_arr)<4:
                print(review)
                reviews_arr.append(review)            
    reviewss.append(" ".join(reviews_arr))

    description = driver.find_element_by_class_name("hp-description").text
    unwanted = driver.find_element_by_class_name("hp_desc_main_content").text
    #print(unwanted)
    #print('-------------------------------------')
    facilities = description.replace(unwanted,"").split("\n")
    #print(facilities)
    facs_arr =[]
    for f in facilities:
        if "Good for solo travellers" in f:
            pass
        else:
            facs_arr.append(f)

    facs = " ".join(facs_arr)
    facilitiess.append(facs)

```

```python
all_df = pd.DataFrame(columns=["Title","Location","City","Facilities","Reviews","Link","Price"])
all_df['Title'] = titles
all_df['City'] = cities
all_df['Facilities'] = facilitiess
all_df['Reviews'] = reviewss
all_df['Location'] = locations
all_df['Price'] = prices
all_df["Link"] = hrefs
all_df.to_excel('Booking1.xls')
all_df
```
