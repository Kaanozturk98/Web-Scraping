```python
from selenium import webdriver
import time
import pandas as pd
import operator
```

```python
chrome_path = r"C:\Users\saxou\Desktop\PYTHON HACK\Web Scrapping\chromedriver_win32 (1)\chromedriver.exe"
driver = webdriver.Chrome(chrome_path)
```

```python
file2 = open('models.txt','r')
model_names = set([x.replace('\n','') for x in file2.readlines()])
brands_txt = set(['google','huawei','nokia','samsung','ulefone','meizu','apple','wileyfox','wiko','motorola','xiaomi','cat','crosscall','doro','alcatel','asus','fairphone','land-rover','ruggear','gigaset','honor','lg','oneplus','oppo','ruggear','sony','renewd'])
colors_txt = set(['space grey','prism','green','gold','purple','ceramic black','prism white','space gray','goud','rose gold','midnight green','keramisch','silver','wit','zilver','rose gold','grijs','rood','black','yellow','white','red','magenta','cyan','zwart','groen','blauw','paars','rosé goud','geel','oranje','brons','bruin','roze'])
```

```python
driver.get("https://www.bol.com/nl/l/smartphones/N/4010/?promo=smartphones_305_type-besturingssysteem_4_alle-smartphones")
```

```python
titles=[]
prices=[]
hrefs=[]
brands=[]
storages=[]
colors=[]
models=[]
competitor=[]

i=0
while i<100:
    products_block = driver.find_element_by_class_name("product-list")
    products = products_block.find_elements_by_class_name("product-item--row")
    print(i+1)
    try:      
        for product in products:
            name = product.find_element_by_class_name("product-title")
            href = name.get_attribute("href")
            hrefs.append(href)

            title = name.text
            titles.append(title)
            
            try:
                price = product.find_element_by_class_name("promo-price").text
                #print("€"+price.replace("\n","."))
                prices.append("€"+price.replace("\n","."))
            except:
                print("para patladi")
                prices.append("-")
            
            specs_PE = product.find_element_by_class_name("product-small-specs")
            specs = specs_PE.find_elements_by_css_selector("li")
            
            #print(specs_PE.text)
            
            if "GB" in specs_PE.text:
                for spec in specs:
                    spec = spec.text
                    if "GB" in spec:
                        spec_arr = spec.split()
                        #print(spec_arr[0]+spec_arr[1])
                        storages.append(spec_arr[0]+spec_arr[1])
            elif "TB" in specs_PE.text:
                for spec in specs:
                    spec = spec.text
                    if "TB" in spec:
                        spec_arr = spec.split()
                        #print(spec_arr[0]+spec_arr[1])
                        storages.append(spec_arr[0]+spec_arr[1])
            elif "MB" in specs_PE.text:
                for spec in specs:
                    spec = spec.text
                    if "MB" in spec:
                        spec_arr = spec.split()
                        #print(spec_arr[0]+spec_arr[1])
                        storages.append(spec_arr[0]+spec_arr[1])
            else:
                storages.append("-")

            try:
                brand = product.find_element_by_class_name('product-creator').text
                #print(brand)
                brands.append(brand)
            except:
                print("marka patladi")
                brands.append("-")
                
            name = title.lower()
            #print(name)

            try:
                tmp = {}
                for clr in colors_txt:
                    #print(clr)
                    if clr in name:
                        tmp[clr] = len(clr)
                colors.append(max(tmp.items(), key=operator.itemgetter(1))[0])
                #print(max(tmp.items(), key=operator.itemgetter(1))[0])
            except:
                colors.append("-")

            try:
                tmp = {}
                for mdl in model_names:
                    #print(mdl)
                    if mdl.lower() in name:
                        tmp[mdl] = len(mdl)
                models.append(max(tmp.items(), key=operator.itemgetter(1))[0])
                #print(max(tmp.items(), key=operator.itemgetter(1))[0])
            except:
                models.append("-")
                
            competitor.append("BOL")
            
        driver.find_element_by_class_name('pagination__controls--next').click()
        time.sleep(4)      
        i+=1
        """
        print(len(storages))
        print(len(brands))
        print(len(hrefs))
        print(len(prices))
        """
    except:
        print("OLMADI")
        break
```

```python
eans=[]
for href in hrefs:
    driver.get(href)
    a = driver.find_element_by_class_name("specs__list").text.split("\n")
    i=0
    for i,b in enumerate(a):
        if b == "EAN":
            eans.append(a[i+1]) 
```

```python
print(len(storages))
print(len(models))
print(len(colors))
print(len(brands))
print(len(hrefs))
print(len(prices))
print(len(eans))
```

```python
all_df = pd.DataFrame(columns=['Brand','Price','Storage',"Link","Model","Color","EAN","Competitor"])
all_df['Model'] = models
all_df['Brand'] = brands
all_df['Price'] = prices
all_df['Color'] = colors
all_df['Storage'] = storages
all_df['Link'] = hrefs
all_df['Competitor'] = competitor
all_df['EAN']=eans
all_df.to_excel('bol1.xls')
all_df
```
