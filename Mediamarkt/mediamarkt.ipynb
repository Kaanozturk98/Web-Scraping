```python
from  selenium import webdriver
import time
import pandas as pd
import operator
```

```python
chrome_path = r"C:\Users\saxou\Desktop\PYTHON HACK\Web Scrapping\chromedriver_win32 (1)\chromedriver.exe"
driver = webdriver.Chrome(chrome_path)
```

```python
driver.get("https://www.mediamarkt.nl/nl/category/_smartphones-483222.html?searchParams=&sort=&view=PRODUCTLIST&page=")
#time.sleep(10)
```

```python
file2 = open('models.txt','r')
model_names = set([x.replace('\n','') for x in file2.readlines()])
brands_txt = set(['google','huawei','nokia','samsung','ulefone','meizu','apple','wileyfox','wiko','motorola','xiaomi','cat','crosscall','doro','alcatel','asus','fairphone','land-rover','ruggear','gigaset','honor','lg','oneplus','oppo','ruggear','sony','renewd'])
colors_txt = set(['space grey','prism','green','gold','purple','ceramic black','prism white','space gray','goud','rose gold','midnight green','keramisch','silver','wit','zilver','rose gold','grijs','rood','black','yellow','white','red','magenta','cyan','zwart','groen','blauw','paars','rosé goud','geel','oranje','brons','bruin','roze'])
```

```python
driver.find_element_by_class_name('cookie-info-accept-button').click()
```

```python
brands=[]
models=[]
colors=[]
prices=[]
storages=[]
hrefs=[]
eans=[]
competitor=[]

i=0
while i<20:
    try:
        products = driver.find_elements_by_class_name("product-wrapper")
        title_2 = []

        for elem in products:
            title_2.append(elem.find_element_by_css_selector("h2"))
            
            name = elem.find_element_by_css_selector("h2").text.lower()
            print(name)
            try:
                tmp = {}
                for clr in colors_txt:
                    #print(clr)
                    if clr in name:
                        tmp[clr] = len(clr)
                colors.append(max(tmp.items(), key=operator.itemgetter(1))[0])
                #print(max(tmp.items(), key=operator.itemgetter(1))[0])
            except:
                colors.append("-")

            try:
                tmp = {}
                for mdl in model_names:
                    #print(mdl)
                    if mdl.lower() in name:
                        tmp[mdl] = len(mdl)
                models.append(max(tmp.items(), key=operator.itemgetter(1))[0])
                print(max(tmp.items(), key=operator.itemgetter(1))[0])
            except:
                models.append("-")

        for elem in title_2:
            elem = elem.text
            elem = elem.split()
            #print(elem)
        
        for elem in title_2:
            elem = elem.text
            #print(elem)
            elem_splt = elem.split()
            brands.append(elem_splt[0])
        
        for elem in products:
            a=elem.find_elements_by_class_name("product-details")
            for i,b in enumerate(a):
                arr = b.text.split()
                #print(arr)                
                if "Geheugencapaciteit:" in arr:
                    for el in arr:
                        if "Geheugencapaciteit:" in el:
                            #print(arr[arr.index(el)+1])
                            storages.append(arr[arr.index(el)+1]+arr[arr.index(el)+2])
                else:
                    arr=title_2[i].text.split()
                    for el in arr:
                        if "GB" in el:
                            storages.append(arr[arr.index(el)-1]+arr[arr.index(el)])

        products = driver.find_element_by_class_name("products-list")
        ps = products.find_elements_by_class_name("small")

        spans=[]

        for p in ps:
            spans.append(p.find_elements_by_css_selector("span"))

        for p in ps:
            span = p.find_elements_by_css_selector("span")
            price = []
            for spn in span:
                price.append(spn.text)
            price = ''.join(price)
            prices.append(price)
            
        contents = products.find_elements_by_class_name("content")
        h2s=[]
        for el in contents:
            h2s.append(el.find_elements_by_css_selector("h2"))

        for h2 in h2s:
            for x in h2:
                a = x.find_element_by_css_selector("a")
                href = a.get_attribute("href")
                hrefs.append(href)
                href_arr = href.replace("-"," ").replace("."," ").split()
                eans.append(href_arr[len(href_arr)-2])
                competitor.append("MediaMarkt")
                
        driver.find_element_by_class_name("pagination-next").click()
        time.sleep(6)
        i+=1
    except:
        break
```

```python
all_df = pd.DataFrame(columns=['Brand','Price','Storage',"Link","Color","Model","EAN","Competitor"])
all_df['Model'] = models
all_df['Brand'] = brands
all_df['Price'] = prices
all_df['Color'] = colors
all_df['Storage'] = storages
all_df['Link'] = hrefs
all_df["EAN"] = eans
all_df["Competitor"]=competitor
all_df.to_excel('mediamarkt.xls')
all_df
```
