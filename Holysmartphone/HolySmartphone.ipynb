```python
from selenium import webdriver
import time
import pandas as pd
import operator
```

```python
chrome_path = r"C:\Users\saxou\Desktop\PYTHON HACK\Web Scrapping\chromedriver_win32 (1)\chromedriver.exe"
driver = webdriver.Chrome(chrome_path)
```

```python
file2 = open('models.txt','r')
model_names = set([x.replace('\n','') for x in file2.readlines()])
brands_txt = set(['google','huawei','nokia','samsung','ulefone','meizu','apple','wileyfox','wiko','motorola','xiaomi','cat','crosscall','doro','alcatel','asus','fairphone','land-rover','ruggear','gigaset','honor','lg','oneplus','oppo','ruggear','sony','renewd'])
colors_txt = set(['space grey','prism','green','gold','purple','ceramic black','prism white','space gray','goud','rose gold','midnight green','keramisch','silver','wit','zilver','rose gold','grijs','rood','black','yellow','white','red','magenta','cyan','zwart','groen','blauw','paars','rosé goud','geel','oranje','brons','bruin','roze'])
```

```python
driver.get("https://holysmartphone.nl/smartphones")
```

```python
time.sleep(3)
SCROLL_PAUSE_TIME = 0.5
try:
    driver.find_element_by_xpath('//*[@id="offcanvas-container"]/div[4]/div/div/a').click()
except:
    pass    

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
products_block = driver.find_element_by_class_name("products-block")
products = products_block.find_elements_by_class_name("product-block")

titles=[]
titles_arr=[]
prices=[]
hrefs=[]
colors=[]
models=[]
competitor=[]

for product in products:
    name = product.find_element_by_class_name("name")
    title = name.text
    if "GB" in title:
        pass
    elif "TB" in title:
        pass
    else:
        products.remove(product)
    
for product in products:
    name = product.find_element_by_class_name("name")
    href = name.find_element_by_css_selector("a").get_attribute("href")
    hrefs.append(href)
    
    title = name.find_element_by_css_selector("a").text
    titles.append(title)
    
    title_arr = title.split()
    titles_arr.append(title_arr)

    price = product.find_element_by_class_name("price").text
    price_arr = price.split()
    if len(price_arr)==2:
        price = price_arr[1]
        prices.append(price)
    else:
        price = price_arr[0]
        prices.append(price)

for title in titles:
    title=title.lower()
    tmp = {}
    for clr in colors_txt:
        #print(clr)
        if clr in title:
            tmp[clr] = len(clr)
    colors.append(max(tmp.items(), key=operator.itemgetter(1))[0])
    #print(max(tmp.items(), key=operator.itemgetter(1))[0])
    
for title in titles:
    title=title.lower()
    tmp = {}
    for mdl in model_names:
        #print(mdl)
        if mdl.lower() in title:
            tmp[mdl] = len(mdl)
    try:
        models.append(max(tmp.items(), key=operator.itemgetter(1))[0])
        #print("x")
    except:
        #print("y")
        models.append("-")
        
    competitor.append("HolySmartphone")
```

```python
brands=[]
storages=[]

for elem in titles_arr:
    if "Iphone" in elem:
        brands.append("Apple")
    else:
        brands.append(elem[0])

for elem in titles_arr:
    for el in elem:
        if "GB" in el:
            storages.append(elem[elem.index(el)])
        if "TB" in el:
            storages.append(elem[elem.index(el)])
```

```python
all_df = pd.DataFrame(columns=['Brand','Price','Storage',"Link","Color","Model","Competitor"])
all_df['Model'] = models
all_df['Brand'] = brands
all_df['Price'] = prices
all_df['Color'] = colors
all_df['Storage'] = storages
all_df['Link'] = hrefs
all_df['Competitor']=competitor
all_df.to_excel('holysmartphone.xls')
all_df
```
