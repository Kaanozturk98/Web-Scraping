```python
from  selenium import webdriver
import time
import pandas as pd
```

```python
chrome_path = r"C:\Users\saxou\Desktop\PYTHON HACK\Web Scrapping\chromedriver_win32 (1)\chromedriver.exe"
driver = webdriver.Chrome(chrome_path)
```

```python
driver.get("https://www.mobiphone.nl/categorie/mobiele-telefonie?orderby=price")
```

```python
titles=[]
prices=[]
hrefs=[]
brands=[]
storages=[]



i=0
while i<25:
    products_block = driver.find_element_by_class_name("products")
    products = products_block.find_elements_by_class_name("product-small")
    print(i+1)
    try:    
        for product in products:
            name = product.find_element_by_class_name("product-title").text
            href_PE = product.find_element_by_css_selector("a")

            try:
                #print(name)
                titles.append(name)
            except:
                titles.append("-")

            try:
                href = href_PE.get_attribute("href")
                #print(href)
                hrefs.append(href)
            except:
                hrefs.append("-")

            try:
                price = product.find_element_by_class_name("amount").text
                prices.append(price)
                #print(price)
            except:
                prices.append("-")


            title_arr = name.split()
            brands.append(title_arr[0])
            #print(title_arr[0])

            if "GB" in name:
                for el in title_arr:
                    if "GB" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])
            elif "gb" in name:
                for el in title_arr:
                    if "gb" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])                       
            elif "Gb" in name:
                for el in title_arr:
                    if "Gb" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])

            elif "TB" in name:
                for el in title_arr:
                    if "TB" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])
            elif "tb" in name:
                for el in title_arr:
                    if "tb" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])                      
            elif "Tb" in name:
                for el in title_arr:
                    if "Tb" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])
                        
            elif "MB" in name:
                for el in title_arr:
                    if "MB" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])
            elif "mb" in name:
                for el in title_arr:
                    if "mb" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])
            elif "Mb" in name:
                for el in title_arr:
                    if "Mb" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])


            else:
                #print("bok")
                storages.append("-")
                
        driver.find_element_by_class_name('next').click()

        time.sleep(4)

        i+=1
        print(len(storages))
        print(len(brands))
        print(len(hrefs))
        print(len(prices))

    except:
        print("OLMADI")
        break
```

```python
print(prices)
print(brands)
print(titles)
print(hrefs)
print(storages)
```

```python
all_df = pd.DataFrame(columns=['Brand','Price','Storage',"Link"])
#all_df['Model'] = models
all_df['Brand'] = brands
all_df['Price'] = prices
#all_df['Color'] = colors
all_df['Storage'] = storages
all_df['Link'] = hrefs
all_df.to_excel('mobiphone.xls')
all_df
```
