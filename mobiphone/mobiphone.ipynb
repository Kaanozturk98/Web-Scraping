```python
from  selenium import webdriver
import time
import pandas as pd
import operator
```

```python
chrome_path = r"C:\Users\saxou\Desktop\PYTHON HACK\Web Scrapping\chromedriver_win32 (1)\chromedriver.exe"
driver = webdriver.Chrome(chrome_path)
```

```python
file2 = open('models.txt','r')
model_names = set([x.replace('\n','') for x in file2.readlines()])
brands_txt = set(['google','huawei','nokia','samsung','ulefone','meizu','apple','wileyfox','wiko','motorola','xiaomi','cat','crosscall','doro','alcatel','asus','fairphone','land-rover','ruggear','gigaset','honor','lg','oneplus','oppo','ruggear','sony','renewd'])
colors_txt = set(['space grey','prism','green','gold','purple','ceramic black','prism white','space gray','goud','rose gold','midnight green','keramisch','silver','wit','zilver','rose gold','grijs','rood','black','yellow','white','red','magenta','cyan','zwart','groen','blauw','paars','rosé goud','geel','oranje','brons','bruin','roze'])
```

```python
driver.get("https://www.mobiphone.nl/categorie/mobiele-telefonie?orderby=price")
```

```python
titles=[]
prices=[]
hrefs=[]
brands=[]
storages=[]
colors=[]
models=[]
competitor=[]



i=0
while i<25:
    products_block = driver.find_element_by_class_name("products")
    products = products_block.find_elements_by_class_name("col")
    print(len(products))
    print(i+1)
    try:    
        for product in products:
            name = product.find_element_by_class_name("product-title").text
            href_PE = product.find_element_by_css_selector("a")

            try:
                print(name)
                titles.append(name)
            except:
                titles.append("-")

            try:
                href = href_PE.get_attribute("href")
                print(href)
                hrefs.append(href)
            except:
                hrefs.append("-")

            try:
                price = product.find_element_by_class_name("amount").text
                prices.append(price)
                print(price)
            except:
                prices.append("-")


            title_arr = name.split()
            brands.append(title_arr[0])
            #print(title_arr[0])

            if "GB" in name:
                for el in title_arr:
                    if "GB" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])
            elif "gb" in name:
                for el in title_arr:
                    if "gb" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])                       
            elif "Gb" in name:
                for el in title_arr:
                    if "Gb" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])

            elif "TB" in name:
                for el in title_arr:
                    if "TB" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])
            elif "tb" in name:
                for el in title_arr:
                    if "tb" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])                      
            elif "Tb" in name:
                for el in title_arr:
                    if "Tb" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])
                        
            elif "MB" in name:
                for el in title_arr:
                    if "MB" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])
            elif "mb" in name:
                for el in title_arr:
                    if "mb" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])
            elif "Mb" in name:
                for el in title_arr:
                    if "Mb" in el:
                        storages.append(title_arr[title_arr.index(el)])
                        #print(title_arr[title_arr.index(el)])
            else:
                #print("bok")
                storages.append("-")
            
            
            name = name.lower()
            print(name)

            try:
                tmp = {}
                for clr in colors_txt:
                    #print(clr)
                    if clr in name:
                        tmp[clr] = len(clr)
                colors.append(max(tmp.items(), key=operator.itemgetter(1))[0])
                #print(max(tmp.items(), key=operator.itemgetter(1))[0])
            except:
                colors.append("-")

            try:
                tmp = {}
                for mdl in model_names:
                    #print(mdl)
                    if mdl.lower() in name:
                        tmp[mdl] = len(mdl)
                models.append(max(tmp.items(), key=operator.itemgetter(1))[0])
                print(max(tmp.items(), key=operator.itemgetter(1))[0])
            except:
                models.append("-")
            
            
            competitor.append("Mobiphone")   
        driver.find_element_by_class_name('next').click()

        time.sleep(4)

        i+=1
        print(i+1)
        print(len(models))
        print(len(colors))
        print(len(hrefs))
        print(len(prices))

    except:
        print("OLMADI")
        break
```

```python
all_df = pd.DataFrame(columns=['Brand','Price','Storage',"Link","Color","Model","Competitor"])

all_df['Model'] = models
all_df['Brand'] = brands
all_df['Price'] = prices
all_df['Color'] = colors
all_df['Storage'] = storages
all_df['Link'] = hrefs
all_df['Competitor']=competitor
all_df.to_excel('mobiphone.xls')
all_df
```
