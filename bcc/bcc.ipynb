```python
file2 = open('models.txt','r')
model_names = set([x.replace('\n','') for x in file2.readlines()])
brands_txt = set(['google','huawei','nokia','samsung','ulefone','meizu','apple','wileyfox','wiko','motorola','xiaomi','cat','crosscall','doro','alcatel','asus','fairphone','land-rover','ruggear','gigaset','honor','lg','oneplus','oppo','ruggear','sony','renewd'])
colors_txt = set(['space grey','prism','green','gold','purple','ceramic black','prism white','space gray','goud','rose gold','midnight green','keramisch','silver','wit','zilver','rose gold','grijs','rood','black','yellow','white','red','magenta','cyan','zwart','groen','blauw','paars','ros√© goud','geel','oranje','brons','bruin','roze'])
```

```python
try:
    driver.find_element_by_xpath('//*[@id="cookiewallmodal"]/div/div/div/button[1]').click()
except:
    pass
```

```python
all_df = pd.DataFrame(columns=['Brand','Price','Storage',"Link","EAN","Color","Model","Competitor"])
```

```python
hrefs=[]
titles=[]

product_list = driver.find_element_by_class_name("productlist")
products = product_list.find_elements_by_class_name("lister-product")
for product in products:
    href = product.find_element_by_class_name("lister-product__titlelink").get_attribute("href")
    #print(href)
    hrefs.append(href)
    title = product.find_element_by_css_selector("h2").text
    print(title)
    titles.append(title)

        
driver.get("https://www.bcc.nl/navigatie-en-telefoon/mobiele-telefoon/smartphone?search=smartphone&index=18")
time.sleep(2)

i=0
while i<25:
    try:
        product_list = driver.find_element_by_class_name("productlist")
        products = product_list.find_elements_by_class_name("lister-product")
        for product in products:
            href = product.find_element_by_class_name("lister-product__titlelink").get_attribute("href")
            #print(href)
            hrefs.append(href)
            title = product.find_element_by_css_selector("h2").text
            print(title)
            titles.append(title)

        driver.find_element_by_xpath('//*[@id="lister-content"]/div/div[3]/div[1]/div/nav/ul/li[6]/a').click()
        time.sleep(2)
        i+=1
    except:
        break
```

```python
len(hrefs)
```

```python
models=[]
brands=[]
for title in titles:
    title=title.lower()
    #print(titles)
    tmp1 = {}
    tmp2 = {}
    for mdl in model_names:
        #print(mdl)
        if mdl.lower() in title:
            tmp1[mdl] = len(mdl)
    models.append(max(tmp1.items(), key=operator.itemgetter(1))[0])
        
    for brand in brands_txt:
        if brand in title:
            tmp2[brand] = len(brand)
    brands.append(max(tmp2.items(), key=operator.itemgetter(1))[0])
```

```python
i=0
for i,href in enumerate(hrefs):
    driver.get(href)
    time.sleep(1)
    
    tmp={}
    
    price = driver.find_element_by_class_name("priceblock").text
    specs = driver.find_element_by_class_name("productspecifications")
    t_bodies = specs.find_elements_by_css_selector("tbody")

    price=price.split()
    #print(price[0]+price[1])
    tmp['Price']=price[0]+price[1]

    storage = t_bodies[2].find_element_by_css_selector("td").text
    #print(storage)
    tmp['Storage']=storage

    ean = t_bodies[len(t_bodies)-1].find_element_by_css_selector("td").text
    #ean
    tmp['EAN']=ean

    for t_body in t_bodies:
        spec_title = t_body.find_element_by_css_selector("tr").text
        if "Design" in spec_title:
            color = t_body.find_elements_by_css_selector("td")[1].text
            #print(color)
            tmp['Color']=color
            
    tmp['Model']=models[i]
    tmp['Brand']=brands[i]
    tmp['Link']=href
    tmp["Competitor"]="BCC"
    i+=1
        
    all_df = all_df.append(tmp,ignore_index=True)
```

```python
all_df.to_excel('bcc.xls')
all_df
```
