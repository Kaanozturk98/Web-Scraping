```python
i=1
for i,link in enumerate(links_123):
    while True:
        start = timeit.default_timer()
        print(i)
        tmp = {}
        driver.get(link)
        tmp['Link']=link

        #####
        SCROLL_PAUSE_TIME = 0.1 
        # Get scroll height
        last_height = driver.execute_script("return document.body.scrollHeight")

        while True:
            # Scroll down to bottom
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            # Wait to load page
            time.sleep(SCROLL_PAUSE_TIME)

            # Calculate new scroll height and compare with last scroll height
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height       
        #####

        #Price
        try:
            try:
                price = driver.find_element_by_class_name("bui-price-display__value").text
                #print(price)
                tmp['Price for night']=price
            except:
                try:
                    WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.CLASS_NAME,'availability_price'))) 
                    price_PE = driver.find_element_by_class_name("availability_price")
                    price = price_PE.find_element_by_css_selector("span").text
                    tmp['Price for night']=price
                except:
                    tmp["Price for night"]="-"
            break                   
        except:
            continue
                
            

    #Title and Type
    title = driver.find_element_by_class_name("hp__hotel-name")
    #print(title.text)
    Type = title.find_element_by_css_selector("span").text
    #tmp['Type']=Type
    #print(Type)
    if Type in title.text:
        name = title.text.replace(Type,"",1).replace(" ","",1)
        tmp['Name of Hotel']=name
    #print(title.text.replace(Type,"",1).replace(" ","",1))

    #Location and City
    location = driver.find_element_by_class_name("hp_address_subtitle").text
    #print(location)
    tmp['Adress']=location
    city = location.split()[len(location.split())-2].replace(",","")
    #print(city)
    tmp['City']=city

    #Star Track
    try:
        star_track = driver.find_element_by_class_name("star_track").get_attribute("title")        
        if "1-star" in star_track:
            tmp['Number of Stars']="1 star"
        elif "2-star" in star_track:
            tmp['Number of Stars']="2 star"
        elif "3-star" in star_track:
            tmp['Number of Stars']="3 star"
        elif "4-star" in star_track:
            tmp['Number of Stars']="4 star"
        elif "5-star" in star_track:
            tmp['Number of Stars']="5 star"
        else:
            tmp['Number of Stars']="-"
    except:
        tmp['Number of Stars']="-"

    #nearby skies and beaches
    try:
        nearby_beaches = driver.find_element_by_class_name("hp-closest-beaches").text
        tmp['Beaches']=nearby_beaches.lower().replace("popular beaches nearby","").replace("\n","",1)
    except:
        pass
    try:
        nearby_beaches = driver.find_element_by_class_name("hp-closest-skilifts").text
        tmp['Mountains']=nearby_beaches.lower().replace("nearby ski lifts:","").replace("\n","",1)
    except:
        pass

    headers=[]
    factis = driver.find_elements_by_class_name("facilitiesChecklistSection")
    for fact in factis:
        fact_title = fact.find_element_by_css_selector("h5").text
        headers.append(fact_title.lower())

    if any("business facilities" in s for s in headers):
        tmp["Working Away"]="1"
    else:
        tmp["Working Away"]="0"

    ttd_num = 0
    factis = driver.find_elements_by_class_name("facilitiesChecklistSection")
    for fact in factis:
        fact_title = fact.find_element_by_css_selector("h5").text
        #print(fact_title)
        if "Ski" in fact_title:
            act_num=len(fact.find_elements_by_css_selector("li"))
            ttd_num+=act_num        
        if "Entertainment & Family Services" in fact_title:
            act_num=len(fact.find_elements_by_css_selector("li"))
            ttd_num+=act_num
        if "Pool and Spa" in fact_title:
            act_num=len(fact.find_elements_by_css_selector("li"))
            ttd_num+=act_num
        if "Health & Wellness Facilities" in fact_title:
            act_num=len(fact.find_elements_by_css_selector("li"))
            ttd_num+=act_num
        if "Activities" in fact_title:
            act_num=len(fact.find_elements_by_css_selector("li"))
            ttd_num+=act_num      
    tmp["Things to do"]=ttd_num

    name=name.lower()
    tmp1={}
    for chn in chains_names:
        if chn.lower() in name:
            tmp1[chn] = len(chn)
    try:
        print(max(tmp1.items(), key=operator.itemgetter(1))[0])
        tmp['Chain Y/N']="Yes"
    except:
        tmp['Chain Y/N']="No"

    print(tmp)
    all_df = all_df.append(tmp,ignore_index=True)
    i+=1

    stop = timeit.default_timer()
    execution_time = stop - start
    print(execution_time)
    print("------------------------------------")
```
